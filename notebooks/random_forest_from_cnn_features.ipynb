##Imports + Load CNN Features

```python
import os
import numpy as np

ART_DIR = os.path.abspath(os.path.join( "artifacts")) 
print("Artifacts dir:", ART_DIR)

F_train = np.load(os.path.join(ART_DIR, "F_train.npy"))
F_val   = np.load(os.path.join(ART_DIR, "F_val.npy"))
F_test  = np.load(os.path.join(ART_DIR, "F_test.npy"))

y_train = np.load(os.path.join(ART_DIR, "y_train.npy"))
y_val   = np.load(os.path.join(ART_DIR, "y_val.npy"))
y_test  = np.load(os.path.join(ART_DIR, "y_test.npy"))

print("Loaded shapes:", F_train.shape, F_val.shape, F_test.shape)
print("Label counts (train/val/test):",
      np.bincount(y_train), np.bincount(y_val), np.bincount(y_test))
```

##Train Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score

rf = RandomForestClassifier(
    n_estimators=500,
    random_state=42,
    class_weight="balanced",
    n_jobs =-1
)

rf.fit(F_train, y_train)
val_pred= rf.predict(F_val)

print("Val acc:", accuracy_score(y_val, val_pred))
print("Val F1 (macro):", f1_score(y_val, val_pred, average ="macro"))
print("Val F1 (tremor =1)", f1_score(y_val, val_pred, pos_label=1))
print("Val confusion:", confusion_matrix(y_val, val_pred))
```

##Evaluation

```python
pred_test= rf.predict(F_test)

print("Test accuracy:", accuracy_score(y_test, pred_test))
print(classification_report(y_test, pred_test))
print("Confusion matrix:\n", confusion_matrix(y_test, pred_test))

print("Test acc:", accuracy_score(y_test, pred_test))
print("Test F1 (macro):", f1_score(y_test, pred_test, average ="macro"))
print("Test F1 (tremor =1)", f1_score(y_test, pred_test, pos_label=1))
print("Test confusion:", confusion_matrix(y_test, pred_test))
```
